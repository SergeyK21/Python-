### 1 Задание Числа Фибоначчи ###

Напишите функцию, которая возвращает список, содержащий числа Фибоначчи, в количестве, переданном в функцию. 

```python
def get_fib_numbers(n: int)->list:
    n=int(n)
    assert n > 0
    list_fib=[0]
    for i in range(n):
        if len(list_fib) >= 2:
            list_fib.append(list_fib[-1]+list_fib[-2])
        else:
            list_fib.append(1)
    return list_fib
get_fib_numbers(10)
```

### Задача 3. FizzBuzz  ###
Выведите числа согласно игре FizzBuzz. 

Суть игры: пользователь вводит максимальное число, и после этого программа выводит значения согласно следующим правилам:

Начинаем с единицы и последовательно увеличиваем значение на один.
Когда число делится на три, оно заменяется на fizz, если число делится на пять, то произносится buzz.
Числа, делящиеся на три и пять одновременно, заменяются на fizz buzz.
При достижении максимального числа мы выводим число согласно правилам и завершаем программу. 

```python
N = int(input('Введи максимальное число N = ')) + 1
for i in range(N):
    if (i % 3 == 0) and (i % 5 == 0) and i != 0:
        print('FizzBuzz')
    elif (i % 5 == 0) and (i != 0):
        print('buzz')
    elif (i % 3 == 0) and (i != 0):
        print('fizz')
    else:
        print(i)
```

### Задача 4. Система расчёта ###
Организуйте систему расчёта статистик успеваемости студентов.

Для начала вы вводите список предметов, после — список фамилий и имён студентов. После для каждого студента выводите оценку в диапазоне от 1 до 10. У вас 10-балльная система оценок.

Когда сбор данных будет завершён, выведите студента с лучшей средней успеваемостью. И средний балл этого студента по предметам.

```python
stat = {'subject':[],'name':[],'firstname':[], 'estimation':[]}

while True:
    subject = input('Введите Предметы ')
    if subject == '' or subject == 'stop':
        break
    stat['subject'].append(subject)

assert len(stat['subject']) !=0, 'Нет предметов'

while True:
    name = input('Введите Имя студента ')
    if name == 'stop' or name == '':
        break
    firstname = input('Введите Фамилию студента ')
    stat['name'].append(name)
    stat['firstname'].append(firstname)

assert len(stat['name']) !=0, 'Нет учеников'

for i in range(len(stat['name'])):
    list_estimation = []
    j = 0;
    while True and j < len(stat['subject']):
        try:
            estimation = int(input(f'Введи оценку по {stat["subject"][j]}\nДля {stat["firstname"][i]} {stat["name"][i]}: '))
            assert (estimation > 1) and (estimation < 10) , "ajkshaj"
            j+=1
            list_estimation.append(estimation)
        except Exception:
            print('\nНеверный ввод.\n')
    stat['estimation'].append(list_estimation)

max = sum(stat['estimation'][0])/len(stat['estimation'][0])
index_max = 0
for i in range(1, len(stat['name'])):
    s = sum(stat['estimation'][i])/len(stat['estimation'][i])
    if s > max:
        max = s
        index_max = i

print(f'Студент с лучшей успеваемостью\n{stat["name"][index_max]} {stat["firstname"][index_max]} {max:.2f}')
```

###  Задача 6. Простые числа ###
Введите максимальное число. Выведите все простые числа, которые не превосходят это число. Если само число является простым, то выведите и его. 

При решении используйте метод «Решето Эратосфена». При выводе выводите в строке числа, принадлежащие одному десятку. 


```python
n=int(input('Введите число: '))
sieve = [i for i in range(n)]
sieve[1] = 0
for i in range(2,n):
    if sieve[i]!=0:
        j=i*2
        while j < n:
            sieve[j] = 0
            j+=i
result = [i for i in sieve if i !=0]
print(result)
```

### Задача 7. «Война и мир» ###
Вам дан файл war_and_peace.txt с романом Льва Толстого «Война и мир».

Прочитайте файл и посчитайте статистику по буквам и символам (string.printable, исключая буквы латинского алфавита, но включая буквы кириллического). Необходимо посчитать отношения количества этих букв к количеству всех символов во всём тексте.

Выведите 10 первых предложений, содержащих имя Андрей (формы «Андрея», «Андрею» и другие вариации). Для упрощения будем считать, что предложение начинается с окончания предыдущего «. » и заканчивается точкой «.».

```python
with open('war_and_peace.txt', 'r') as file:
    str_w = file.read()

ABC_ru = []
count_simvol = 0
for i in str_w:
    if i != '\n' and i != '\t':
        count_simvol += 1
    else:
        continue
    if i.isalpha() and (((ord(i) >= ord('а') and ord(i) >= ord('я')) or (ord(i) >= ord('А') and ord(i) >= ord('Я'))
                                    or ord(i) == ord('Ё') or ord(i) == ord('ё'))):
        ABC_ru.append(i.upper())
alf_set = list(set(ABC_ru))
alf_set.sort()

for i in alf_set:
    print(f"буква '{i}' встретилась {ABC_ru.count(i)} раз, а всего в тексте {count_simvol} символов, значит, частота '{i}' — "
          f"{ABC_ru.count(i)}/{count_simvol} = {ABC_ru.count(i)/count_simvol:.6f}")

list_str_w = str_w.split(sep='.')

list_name = []

for i in list_str_w:
    if i.count("Андр"):
        list_name.append(i)
    if len(list_name) == 10:
        break

for i, temp_lst in enumerate(list_name):
    print(i+1, temp_lst)
```

### Задача 8. Калькулятор ###
Сделайте калькулятор, который поддерживает четыре операции: сложение (+), умножение (×), вычитание (−), деление (÷).

Калькулятор должен поддерживать операции с целыми числами и числами с плавающей точкой, а также должен быть толерантен к пробелам, то есть между операндами и числами может быть неограниченное число пробелов.
Как работает калькулятор
Вы указываете имя файла, содержащего выражения, которые нужно вычислить. Одно выражение — одна строка. Если в выражении есть ошибки, калькулятор запоминает номера строк, в которых они встретились. 

Калькулятор выводит результаты расчётов, номер строки и результат вычисления, а также количество ошибок, номера строк, в которых они встретились, и тип ошибок. 

```python
with open('calc.txt', 'w+', encoding='utf-8') as file:
    file.write("3,1  +    d\n")
    file.write("  4,3 + 5  \n")
    file.write("  4 - 5  \n")
    file.write("  4 * 5  \n")
    file.write("  20 / 5  \n")
    file.write("  21 / 5  \n")
    file.write("  21 / 0  \n")


with open('calc.txt', 'r', encoding='utf-8') as file:
    list_str = file.readlines()
result = []
for i in list_str:
    temp_str = i.replace(' ','')
    temp_str = temp_str.replace('\n', '')
    temp_str = temp_str.replace('\t', '')
    temp_str = temp_str.replace(',', '.')
    try:
        if temp_str.count("+") > 0:
            if temp_str.count("+") > 1:
                raise Exception("В вырвжении не должно быть более одного опператора")
            temp_str = temp_str.split(sep="+")
            temp_digit = []
            for el in temp_str:
                temp_digit.append(float(el))
            result.append(temp_digit[0] + temp_digit[1])
        elif temp_str.count("-"):
            if temp_str.count("-") > 1:
                raise Exception("В вырвжении не должно быть более одного опператора")
            temp_str = temp_str.split(sep="-")
            temp_digit = []
            for el in temp_str:
                temp_digit.append(float(el))
            result.append( temp_digit[0] - temp_digit[1])

        elif temp_str.count("*"):
            if temp_str.count("*") > 1:
                raise Exception("В вырвжении не должно быть более одного опператора")
            temp_str = temp_str.split(sep="*")
            temp_digit = []
            for el in temp_str:
                temp_digit.append(float(el))
            result.append( temp_digit[0] * temp_digit[1])

        elif temp_str.count("/"):
            if temp_str.count("/") > 1:
                raise Exception("В вырвжении не должно быть более одного опператора")
            temp_str = temp_str.split(sep="/")
            temp_digit = []
            for el in temp_str:
                temp_digit.append(float(el))
            #
            # if temp_digit[1] == 0:
            #     raise Exception("На ноль делить нельзя")
            result.append( temp_digit[0] / temp_digit[1])

    except Exception as ex:
        result.append(str(ex))


with open('calc_result.txt', 'w+', encoding='utf-8') as file:
    for i in range(len(list_str)):
        temp=list_str[i].replace('\n','')
        file.write(f"{temp} = {result[i]}\n")
```
